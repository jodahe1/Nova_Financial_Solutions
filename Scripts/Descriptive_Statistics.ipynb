```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data  = pd.read_csv("C:/Users/huawei/Desktop/Nova_Financial_Solutions/Data/raw_analyst_ratings.csv")
```

```python
# Display the first few rows of the dataset
print("First few rows :")
print(data.head())
print()
```

```python
# Check the shape (number of rows and columns) of The dataset
print("raw_analyst_ratings:", data.shape)
```

```python
# Check the data types of each column
print("Data types in raw_analyst_ratings:")
print(data.dtypes)
print()
```

```python
# Descriptive statistics for textual lengths
data['headline_length'] = data['headline'].apply(lambda x: len(x))
headline_stats = data['headline_length'].describe()
print("Descriptive Statistics for Headline Length:")
print(headline_stats)
```

```python
# Count the number of articles per publisher
publisher_counts = data['publisher'].value_counts()
print("\nNumber of Articles per Publisher:")
print(publisher_counts)
```

```python

# Plot a histogram of headline lengths
plt.figure(figsize=(10, 6))
plt.hist(data['headline_length'], bins=20, edgecolor='black')
plt.title('Distribution of Headline Lengths')
plt.xlabel('Headline Length')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plot a bar chart of article counts by publisher
plt.figure(figsize=(10, 6))
publisher_counts.plot(kind='bar')
plt.title('Number of Articles per Publisher')
plt.xlabel('Publisher')
plt.ylabel('Number of Articles')
plt.show()
```

```python

# Analyze publication dates
data['date'] = pd.to_datetime(data['date'].str.split(' ').str[0])
data['day_of_week'] = data['date'].dt.day_name()
```

```python

# Plot article frequency by day of the week
daily_counts = data['day_of_week'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
daily_counts.plot(kind='bar')
plt.title('Article Frequency by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Articles')
plt.show()
```
